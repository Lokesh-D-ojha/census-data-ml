# FIlling worclass with most frequent workclasses

```python
frequentwc = df['workclass'].mode()[0]
df['workclass'] = df['workclass'].replace('?',frequentwc)
print(df['workclass'].head(10))
```

# Filling missing occupation values based on workclass mode

```python
df['occupation'] =df['occupation'].replace('?',np.nan)
for workclass in df['workclass'].unique():
    mode = df.loc[df['workclass'] == workclass, 'occupation'].mode()
    if not mode.empty:
        df.loc[(df['workclass'] == workclass) & (df['occupation'].isna()), 'occupation'] = mode[0]
    else:
        df['workclass'].fillna('Unknown')
df['occupation'] = df['occupation'].fillna('Unknown')
```

# Filling native.country with the country with most frequent occurrence in the native.country column

```python
df['native.country'] = df['native.country'].replace('?', np.nan)
common_country = df['native.country'].mode()[0]
df['native.country'] = df['native.country'].fillna(common_country)
# Verify
print(df['native.country'].isna().sum())
```
